```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
import datetime
```

# Zadanie 1

```python
nasa = pd.read_csv("./course-files/nasa_facebook_statuses.csv", usecols=['status_published','num_reactions','num_comments','num_shares','num_likes'])
nasa["status_published"] = pd.to_datetime(nasa["status_published"])
filter =(nasa['status_published'] >= '2016-5-1') & (nasa['status_published'] < '2016-6-1')
nasa = nasa[filter]
nasa['day'] = nasa.apply(lambda row: row['status_published'].day,axis=1)
nasa_by_day = nasa.groupby(by='day').sum()
nasa_by_day.head()
```

```python
nasa_by_day.plot()
```

```python
nasa_by_day['num_shares'].plot()
```

```python
nasa_by_day[['num_shares','num_comments']].plot()
```

```python
nasa_by_day.plot(y='num_shares')
```

```python
nasa_by_day.plot(y=['num_shares','num_comments'])
```

# Zadanie 4

```python
nasa = pd.read_csv("./course-files/nasa_facebook_statuses.csv",usecols=["status_published","num_reactions","num_comments","num_shares","num_likes"])
nasa["status_published"] = pd.to_datetime(nasa["status_published"])
filter =(nasa['status_published'] >= '2016-3-1') & (nasa['status_published'] < '2016-6-1')
nasa = nasa[filter]
nasa['month'] = nasa.apply(lambda row: row["status_published"].month,axis=1)
nasa_by_month = nasa.groupby(by='month').sum()
nasa_by_month.head()
```

```python
plt.style.use('default')
nasa_by_month['num_likes'].plot(kind='pie',figsize=(5,5),title='NASA Facebook activity')
```

```python
nasa_by_month.plot(kind='pie',figsize=(16,4),title='NASA Facebook activity',subplots=True)
```

```python
nasa_by_month.plot(kind='pie',figsize=(10,10),title='NASA Facebook activity',subplots=True,labels=("March","April","June"), layout=(2,2),legend=False)
```

```python
nasa_by_month.plot(kind='pie',figsize=(10,10),title='NASA Facebook activity',subplots=True,labels=("March","April","June"), 
                   layout=(2,2),autopct="%.0f%%",textprops=dict(color='white'))
```
