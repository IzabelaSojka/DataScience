```python
suppliers = pd.read_csv("./course-files/northwind-mongo-master/suppliers.csv", usecols=['SupplierID','CompanyName','City'],low_memory=False)
products = pd.read_csv("./course-files/northwind-mongo-master/products.csv", usecols=['ProductID','ProductName','SupplierID','CategoryID'],low_memory=False)
categories = pd.read_csv("./course-files/northwind-mongo-master/categories1.csv", usecols=['CategoryID','CategoryName'],index_col = 'CategoryID',low_memory=False)
```

```python
prod_cat = products.merge(categories, left_on='CategoryID', right_index=True)
prod_cat.head()
```

```python
prod_cat_sup = prod_cat.merge(suppliers, left_on='SupplierID', right_on='SupplierID')
prod_cat_sup.head()
```
